```python
#import pandas to create the dataframe
import pandas as pd
import numpy as np
```

```python
#start with first dataset 
#save data in dataframe
fake = pd.read_csv('Fake.csv')
```

```python
#view data 
fake.head()
```

```python
#do the same for real
real =  pd.read_csv('True.csv')
```

```python
real.head()
```

```python
#just need title and text so save those
fake = fake[['title' , 'text']].copy()
real = real[['title' , 'text']].copy()
```

```python
#Add label field to both and rename columns according to prototype to make it easier to use the same code
fake['Label'] = 'fake'
fake.rename(
    columns={"title": "Headline", "text": "Body"},
    inplace=True,
)
fake.head()
```

```python
#Do the same for real
real['Label'] = 'real'
real.rename(
    columns={"title": "Headline", "text": "Body"},
    inplace=True,
)
real.head()
```

```python
#combine these two together
first_df = pd.concat([fake,real] , axis=0)
len(first_df)
```

```python
#import the next dataset
fake_or_real = pd.read_csv('fake_or_real_news.csv')
```

```python
fake_or_real.head()
```

```python
#Again just keep title, text and label and rename fields
fake_or_real = fake_or_real[['title' , 'text','label']].copy()
```

```python
fake_or_real.rename(
    columns={"title": "Headline", "text": "Body" , "label":"Label"},
    inplace=True,
)
fake_or_real.head()
```

```python
#Now because previously label was lowercase, do that here
fake_or_real['Label'] = np.where(fake_or_real['Label'] == 'REAL', 'real', fake_or_real['Label'])
fake_or_real['Label'] = np.where(fake_or_real['Label'] == 'FAKE', 'fake', fake_or_real['Label'])
fake_or_real.head()
```

```python
#concatenate this to the previous dataset
second_df = pd.concat([first_df,fake_or_real] , axis=0)
len(second_df)
```

```python
data = pd.read_csv('data.csv')
```

```python
data.head()
```

```python
#just keep the Headline.Body and Label and change 1 to real and 0 to fake
data = data[['Headline' , 'Body' , 'Label']].copy()
```

```python
data['Label'] = np.where(data['Label'] == 1, 'real', data['Label'])
data['Label'] = np.where(data['Label'] == 0, 'fake', data['Label'])
data.head()
```

```python
#concatenate now
third_df= pd.concat([second_df,data] , axis=0)
len(third_df)
```

```python
#check if real and fake values are similar.
label_counts = third_df['Label'].value_counts()
print(label_counts)
```

```python
#Since it returned some labels as 0, view those
zero_rows = third_df[third_df['Label'] == '0']
print(zero_rows)
```

```python
#delete these rows
third_df = third_df[third_df['Label'] != '0']
```

```python
#Now lets check
label_counts = third_df['Label'].value_counts()
print(label_counts)
```

```python
#Now do the same for the bbc dataset
bbc = pd.read_csv('bbc_news.csv')
```

```python
bbc.head()
```

```python
#shuffle the dataset to get random 
bbc = bbc.sample(frac = 1).reset_index()
bbc.head()
```

```python
bbc_sub = bbc.head(185)
```

```python
#change headers, remove the columns not needed and add a label column
bbc_sub = bbc_sub[['title' , 'description']].copy()
bbc_sub['Label'] = 'real'
bbc_sub.rename(
    columns={"title": "Headline", "description": "Body"},
    inplace=True,
)
bbc_sub.head()
```

```python
#add it to the dataset
df = pd.concat([third_df,bbc_sub] , axis=0)
len(df)
```

```python
#remove empty rows
df.dropna()
```

```python
# check real to fake ratio again
label_counts = df['Label'].value_counts()
print(label_counts)
```

```python
#looks good so shuffle data
df = df.sample(frac = 1).reset_index()
df.head()
```

```python
#remove index
df = df[['Headline' , 'Body' , 'Label']].copy()
df.head()
```

```python
#create a new df with labels and headline and body combined.
new_df = pd.DataFrame()

new_df['Text'] = df['Headline'] + " " + df['Body']

new_df['Label'] = df['Label']

new_df.head()
```

```python
#export it as csv
new_df.to_csv("dataframe.csv",index=False)
```
